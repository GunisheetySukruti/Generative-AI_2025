```python
#run this
import os
from PIL import Image
import torch
from torchvision import transforms

# Load images
image_paths = [os.path.join("/content/drive/MyDrive/19 (1)", img) for img in os.listdir("/content/drive/MyDrive/19 (1)")]
images = [Image.open(img).convert("RGB") for img in image_paths]

# Preprocess images (resize and normalize)
preprocess = transforms.Compose([
    transforms.Resize((512, 512)),
    transforms.ToTensor(),
    transforms.Normalize([0.5], [0.5])
])

# Apply preprocessing
processed_images = [preprocess(img) for img in images]
```

```python
#run this
# Load the Stable Diffusion model
from diffusers import StableDiffusionPipeline
import torch

model_id = "stabilityai/stable-diffusion-2-1"
pipe = StableDiffusionPipeline.from_pretrained(
    model_id, torch_dtype=torch.float32  # Use float32 for CPU
)

# Set the device to CPU
pipe.to("cpu")

# Generate an image
prompt = "A beautiful sunset over a serene mountain landscape"
output = pipe(prompt)

# Save the generated image
image = output.images[0]
image.save("generated_landscape.png")
#to dispaly image
display(image)
```
